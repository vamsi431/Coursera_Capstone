```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print("Fetching venues for : ",name)            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbors',
                             'Neighbors Latitude', 
                             'Neighbors Longitude', 
                             'Venue', 
                             'Venue Latitude', 
                             'Venue Longitude', 
                             'Venue Category']
    
    return(nearby_venues)
```

```python
vis_venues = getNearbyVenues(names=vis['Neighbors'],latitudes=vis['Latitude'],longitudes=vis['Longitude'])
```

```python
vis_onehot=pd.get_dummies(vis_venues[['Venue Category']],prefix='',prefix_sep='')
vis_onehot['Neighbors']=vis_venues['Neighbors']
fixed_columns=[vis_onehot.columns[-1]]+list(vis_onehot.columns[:-1])
vis_onehot=vis_onehot[fixed_columns]
print(vis_onehot.shape)
vis_onehot.head()
```

```python
vizag_grouped=vis_onehot.groupby(["Neighbors"]).mean().reset_index()
print(vizag_grouped.shape)
vizag_grouped.head()
```

```python
len(vizag_grouped[vizag_grouped['Indian Restaurant']>0])
```

```python
vis_restaurant=vizag_grouped[["Neighbors","Indian Restaurant"]]
```

```python
vis_restaurant.head()
```

```python

kclusters=3
vizag_clustering=vis_restaurant.drop(['Neighbors'],1)
kmeans=KMeans(n_clusters=kclusters,random_state=0).fit(vizag_clustering)
kmeans.labels_[0:10]
```

```python
vizag_merge=vis_restaurant.copy()
```

```python
vizag_merge['Cluster Labels']=kmeans.labels_
```

```python
vizag_merge.head()

vizag_merge=vizag_merge.join(vis_venues.set_index('Neighbors'),on='Neighbors')
print(vizag_merge.shape)
vizag_merge.head()
```

```python
print(vizag_merge.shape)
vizag_merge.sort_values(['Cluster Labels'],inplace=True)
vizag_merge.head()
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[17.6868, 83.2185], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(vizag_merge['Neighbors Latitude'], vizag_merge['Neighbors Longitude'], vizag_merge['Neighbors'], vizag_merge['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
